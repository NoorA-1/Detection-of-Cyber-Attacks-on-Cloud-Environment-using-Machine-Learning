# Feature Engineering

```python
df_cleaned = df_cleaned.drop(['UUID', 'dom', 'VMID', 'LAST_POLL'], axis=1)
df_cleaned['Status'] = df_cleaned['Status'].map({'Normal': 0, 'Attack': 1})
df_cleaned.head()
```

```python
correlation_matrix = df_cleaned.corr()

high_corr_threshold = 0.9

high_corr_pairs = correlation_matrix.abs().unstack().sort_values(kind="quicksort", ascending=False)
high_corr_pairs = high_corr_pairs[high_corr_pairs > high_corr_threshold]

high_corr_pairs = high_corr_pairs[high_corr_pairs != 1.0]

high_corr_pairs = high_corr_pairs.drop_duplicates()
high_corr_pairs
```

```python
features_to_drop = [
    'memusable_slope',  # Correlated with memunused_slope
    'memminor_fault_slope',  # Correlated with memunused_slope and memlast_update_slope
    'memlast_update_slope'  # Correlated with memunused_slope
]

df_cleaned = df_cleaned.drop(columns=features_to_drop)

remaining_columns = df_cleaned.columns
remaining_columns
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Split the data into features and target
X = df_cleaned.drop(columns=['Status'])
y = df_cleaned['Status']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred = rf.predict(X_test)

# Calculate accuracy and display classification report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Get feature importance
feature_importances = rf.feature_importances_
important_features = pd.Series(feature_importances, index=X.columns).sort_values(ascending=False)

print(important_features.head(10))
```

```python
top_features = important_features.head(10).index
X_top = df_cleaned[top_features]

X_top.head()
```

# Model Training and Evaluation

```python
X = df_cleaned[top_features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

model_dt = DecisionTreeClassifier()

model_dt.fit(X_train, y_train)
y_pred = model_dt.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model: {model_dt.__class__.__name__}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print("\n")
print("Confusion Matrix:")
# print(confusion_matrix(y_test, y_pred))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier

model_et = ExtraTreesClassifier()
model_et.fit(X_train, y_train)

y_pred = model_et.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model: {model_et.__class__.__name__}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print("\n")
print("Confusion Matrix:")
# print(confusion_matrix(y_test, y_pred))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.show()
```

```python
from sklearn.svm import SVC
model_svm = SVC(random_state=42)

model_svm.fit(X_train, y_train)
y_pred = model_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model: {model_svm.__class__.__name__}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print("\n")
print("Confusion Matrix:")
# print(confusion_matrix(y_test, y_pred))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()

model_knn.fit(X_train, y_train)
y_pred = model_knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model: {model_knn.__class__.__name__}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print("\n")
print("Confusion Matrix:")
# print(confusion_matrix(y_test, y_pred))
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.show()
```
